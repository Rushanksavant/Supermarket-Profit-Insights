```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../input/marketing-data-for-a-supermarket-in-united-states/supermarket_marketing/Supermarket_CustomerMembers.csv')
data.head()
```

# Data visualizaion 

```python
plt.rcParams["figure.figsize"] =(18,6)

plt.subplot(221)
sns.countplot(data.Genre)

plt.subplot(222)
sns.countplot(data.Age)

plt.subplot(223)
sns.countplot(data["Annual Income (k$)"])

plt.subplot(224)
sns.countplot(data["Spending Score (1-100)"])
```

# Knowing maximum and minimum ages of customers

```python
ages= [] # Will contain all the ages values

for i in range(len(data)):
    ages.append(data["Age"][i])
    
    
print("Maximum customer age is: " + str(max(ages)))
print("Minimum customer age is: " + str(min(ages)))
```

# Segregating Customers based on age
* Under 20 years
* 20 to 24 years
* 25 to 29 years
* 30 to 34 years
* 35 to 39 years
* 40 to 44 years
* 45 to 49 years
* 50 to 54 years
* 55 to 60 years
* 60 and above years

```python
data["Age Group"]= 0 # Adding new column to data for Age group info
data.head()
```

```python
for i in range(len(data)):
    
    if data["Age"][i]< 20:
        data["Age Group"][i]= 'Under 20 years'
        
    if 20<= data["Age"][i]<25:
        data["Age Group"][i]= '20 to 24 years'
        
    if 25<= data["Age"][i]<30:
        data["Age Group"][i]= '25 to 29 years'
        
    if 30<= data["Age"][i]<35:
        data["Age Group"][i]= '30 to 34 years'
        
    if 35<= data["Age"][i]<40:
        data["Age Group"][i]= '35 to 39 years'
        
    if 40<= data["Age"][i]<45:
        data["Age Group"][i]= '40 to 44 years'
        
    if 45<= data["Age"][i]<50:
        data["Age Group"][i]= '45 to 49 years'
        
    if 50<= data["Age"][i]<55:
        data["Age Group"][i]= '50 to 54 years'
        
    if 55<= data["Age"][i]<60:
        data["Age Group"][i]= '55 to 60 years'
        
    if 60<= data["Age"][i]:
        data["Age Group"][i]= '60 and above years'
        
display(data)        
```

# Gender-Age Group 

```python
plt.rcParams['figure.figsize'] = (10, 5)
Y = pd.crosstab(data['Age Group'], data['Genre'])
Y.div(Y.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True)
plt.legend(loc="upper right")
plt.show()
```

## Knowing maximum and minimum income of customers 

```python
income= [] # Will contain all the income values

for i in range(len(data)):
    income.append(data["Annual Income (k$)"][i])
    
    
print("Maximum customer age is: " + str(max(income)) + ' k$')
print("Minimum customer age is: " + str(min(income)) + ' k$')
```

# Annual Income based on Age group and Gender

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.scatterplot(y="Age Group",x="Annual Income (k$)",data=data,hue="Genre")
```
