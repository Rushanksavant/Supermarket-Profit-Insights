# Counting number of Clicks(intrest take by people) for all Ads 

```python
ad_rat= {} # Will contain Ad number : number of clicks
for i in range(len(list(data.columns))):
    
    k= 0
    
    for j in range(len(data)):
        
        if data[list(data.columns)[i]][j]== 1:
            k= k+1
     
    ad_rat[f'Ad{i+1}'] = k   
        
```

```python
ad_rat
```

```python
ads= list(ad_rat.keys())
num_click= list(ad_rat.values())

# Set the width and height of the figure
plt.figure(figsize=(20,9))

# Add title
plt.title("Ad number")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=ads, y=num_click)

# Add label for vertical axis
plt.ylabel("Number of clicks")
```

# Percentage performance of ads

```python
for i in range(len(list(data.columns))):
    print(f"Percentage performance of Ad {i+1}: " +str(num_click[i]*100/len(data)) + '%')
    print("\n")
```
